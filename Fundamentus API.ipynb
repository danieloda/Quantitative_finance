```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
r = requests.get('https://www.fundamentus.com.br/resultado.php?setor=4')
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
table = soup.find('table', class_ = 'resultado rowstyle-par colstyle-col no-arrow')
```

```python
cabecalho = []
for columns in table.find_all('thead'):
    cols = columns.find_all('th')
        
for i in range(len(cols)):
    cabecalho.append(cols[i].find('a').text)

index = []
for lines in table.find_all('tbody'):
    line = lines.find_all('tr')

for i in range(len(line)):
    index.append(line[i].find('a').text)

content = []
for elements in table.find_all('tbody'):
    rows = elements.find_all('td')

for i in range(len(line)):
    linha = line[i].text.split()
    content.append(np.array(linha))
    
tabela = pd.DataFrame(data = content,columns = cabecalho, index = index)
tabela = tabela.drop(axis=1, columns = 'Papel')

df = tabela.copy()
#df = df.drop('CEPE6')
df
```

```python
for coluna in ['Cotação','P/L','P/VP','PSR','P/Ativo','P/Cap.Giro','P/Ativ Circ.Liq'
               ,'EV/EBITDA','Liq. Corr.','Dív.Brut/ Patrim.']:
    df[coluna] = df[coluna].str.replace(',','.').astype('float')


for coluna in ['Div.Yield','Mrg Ebit','Mrg. Líq.','ROIC','ROE','Cresc. Rec.5a']:
    df[coluna] = df[coluna].str.replace(',','.')
    df[coluna] = df[coluna].str.replace('.','')
    df[coluna] = df[coluna].str.rstrip('%').astype('float') / 100

df
```

```python
ranking = pd.DataFrame()
ranking['pos'] = range(1,len(df)+1)
ranking['P/L'] = df['P/L'].sort_values(ascending = True).index
ranking['P/VP'] = df['P/VP'].sort_values(ascending = True).index
ranking['PSR'] = df['PSR'].sort_values(ascending = True).index
ranking['ROE'] = df['ROE'].sort_values(ascending = False).index
ranking['Dív.Brut/ Patrim.'] = df['Dív.Brut/ Patrim.'].sort_values(ascending = True).index
ranking['Cresc. Rec.5a'] = df['Cresc. Rec.5a'].sort_values(ascending = False).index
ranking
```

```python
a = ranking.pivot_table(columns='P/L', values='pos')
b = ranking.pivot_table(columns='P/VP', values='pos')
c = ranking.pivot_table(columns='PSR', values='pos')
d = ranking.pivot_table(columns='ROE', values='pos')
e = ranking.pivot_table(columns='Dív.Brut/ Patrim.', values='pos')
f = ranking.pivot_table(columns='Cresc. Rec.5a', values='pos')
t = pd.concat([a,b,c,d,e,f])
rank = t.dropna(axis=1).sum()
lista = list(rank.sort_values(ascending = False).index[:30])
lista
```
